### Data Load

```python
import sys
sys.path.append('/content/drive/MyDrive/Advertising_By_Personality/src/business_model/t5_selling')
print(sys.path)
```

```python
import re
import json
import swifter
import pandas as pd
from tqdm import tqdm
from pshmodule.utils import filemanager as fm
```

```python
df = fm.load('/content/drive/MyDrive/Advertising_By_Personality/data/origin/origin.xlsx')
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df.rename(columns={'분류':'type', '제목':'title', '본문':'content'}, inplace=True)
```

```python
df = df[['type', 'title', 'content']]
```

반 나누기

```python
df1 = df.iloc[:12500]
df2 = df.iloc[12500:]
```

```python
t_list1 = ['원문', 'NT']
t_list2 = ['원문', 'NF']
df1 = df1[df1['type'].isin(t_list1)]
df2 = df2[df2['type'].isin(t_list2)]
```

```python
df1.shape
```

```python
df1.head()
```

```python
df2.head()
```

### Parallel

```python
result1 = []
temp1 = []

for i in df1.iterrows():
  temp1.append([i[1]['type'], i[1]['title'], i[1]['content']])
  if len(temp1) == 2:
    result1.append(temp1)
    temp1 = []
```

```python
result2 = []
temp2 = []

for i in df2.iterrows():
  temp2.append([i[1]['type'], i[1]['title'], i[1]['content']])
  if len(temp2) == 2:
    result2.append(temp2)
    temp2 = []
```

```python
result1[:2]
```

```python
result2[:2]
```

```python
df_result = pd.DataFrame(result1 + result2, columns=['dummy1', 'dummy2'])
```

```python
df_result["type1"] = df_result.dummy1.swifter.apply(lambda x: x[0])
df_result["title1"] = df_result.dummy1.swifter.apply(lambda x: x[1])
df_result["content1"] = df_result.dummy1.swifter.apply(lambda x: x[2])
df_result["type2"] = df_result.dummy2.swifter.apply(lambda x: x[0])
df_result["title2"] = df_result.dummy2.swifter.apply(lambda x: x[1])
df_result["content2"] = df_result.dummy2.swifter.apply(lambda x: x[2])
```

```python
df_result["sent1"] = df_result.title1 + "\\\\" + df_result.content1
df_result["sent2"] = df_result.title2 + "\\\\" + df_result.content2
```

```python
df_result = df_result[['type1', 'sent1', 'type2', 'sent2']]
```

```python
df_result.head()
```

```python
df_result.tail()
```

```python
df_result.shape
```

### emoji

```python
from pshmodule.processing import processing as p
```

```python
nlp = p.Nlp()
```

```python
df_result["sent1"] = df_result.sent1.swifter.apply(nlp.convert_to_python_unicode)
df_result["sent2"] = df_result.sent2.swifter.apply(nlp.convert_to_python_unicode)
```

```python
df_result["sent1"] = df_result.sent1.swifter.apply(nlp.convert_emojis_in_text)
df_result["sent2"] = df_result.sent2.swifter.apply(nlp.convert_emojis_in_text)
```

```python
df_result["sent1"] = df_result.sent1.swifter.apply(lambda x: x.encode("utf-8", "ignore").decode("utf-8"))
df_result["sent2"] = df_result.sent2.swifter.apply(lambda x: x.encode("utf-8", "ignore").decode("utf-8"))
```

```python
df_result.tail()
```

### Save

```python
temp_dict = [{"input": row['sent1'].strip(), "label": row['sent2']} for _, row in df_result.iterrows()]
```

```python
with open('drive/MyDrive/Advertising_By_Personality/data/train_t5_selling.json', 'w', encoding='utf-8') as f:
    for line in temp_dict:
        json_record = json.dumps(line, ensure_ascii=False)
        f.write(json_record + '\n')
```
